```python
import pandas as pd
import numpy as np

df = pd.read_csv('./Datos/creditcard.csv')
output_path = './Datos/creditcard_balanced_view.csv'

# 1. EXTRACCIÓN
print("Cantidad de registros:", df.shape[0])

# 2. TRANSFORMACIÓN
# a. Eliminación de Valores Nulos
if df.isnull().sum().sum() > 0:
    df.fillna(0, inplace=True) # O la estrategia que prefieras

# b. Escala Logarítmica para 'Amount'
# Ayuda a visualizar mejor los montos pequeños vs grandes
df['Amount_Log'] = np.log1p(df['Amount'])

# c. Manejo de Desbalanceo: Se realiza un muestreo para crear un dataset balanceado para visualización
# Creamos un dataset balanceado solo para gráficos específicos donde queremos ver patrones
frauds = df[df['Class'] == 1]
non_frauds = df[df['Class'] == 0].sample(n=len(frauds) * 10, random_state=42) # Ratio 1:10
df_balanced = pd.concat([frauds, non_frauds])

# 3. CARGA (Load)
# Guardamos el dataset procesado para que Tableau lo consuma
df_balanced.to_csv(output_path, index=False)
print("Dataset balanceado guardado en:", output_path)
```
